## Part 1 - Retrieve the data, and examine it 

```python
import requests
import pandas as pd
```

```python
base_url = 'http://linserv1.cims.nyu.edu:10000/films?_page='
```

```python
page = 1
```

```python
dfs = []
while True: 
    url = f'{base_url}{page}'
    res = requests.get(url)
    d = res.json()
    if len(d) == 0: 
        break 
    else:
        dfs.append(pd.DataFrame(d))
        page += 1
```

*When the URL is modified*
- When I modify the url I get different displays of movies. When the number is incremented above 3, nothing is shown and an empty page exists.

## Part 3 - Report

```python
whole_df = pd.concat([movies[0], movies[1], movies[2]])
```

```python
whole_df
```

```python
whole_df.info()
```

```python
whole_df['rt_score'] = pd.to_numeric(whole_df['rt_score'])
```

```python
whole_df.info()
```

```python
whole_df['rt_score'].groupby(whole_df['director']).mean()
```

```python
report_df = pd.DataFrame(whole_df['rt_score'].groupby(whole_df['director']).mean())
```

```python
report_df
```

```python
report_df_add = pd.DataFrame(whole_df['rt_score'].groupby(whole_df['director']).size())
```

```python
final_report = report_df.merge(report_df_add,  how='left', on='director')
```

```python
final_report
```

```python
final_report.rename(columns = {'rt_score_x':'avg_rt_score', 'rt_score_y':'count'})
```

```python
final_report.sort_values(by=['rt_score_x'], ascending=False)
```
